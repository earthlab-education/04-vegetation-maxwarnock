```python
# Compute the difference in NDVI before and after the return of water rights in 2004
ndvi_diff = (
    ndvi_da
        .sel(date=slice('2012', '2022'))
        .mean('date')
        .NDVI
    - ndvi_da
        .sel(date=slice('2001','2011'))
        .mean('date')
        .NDVI
)

ndvi_diff
```

```python
# Plot the difference
(
    ndvi_diff.hvplot(x='x', y='y', cmap='PiYG', geo=True, title="NDVI Changes in Gila River Indian Community \n2012-2022 vs. 2001-2011")
    *
    boundary_gdf.hvplot(geo=True, fill_color=None, line_color='black')
)
```
